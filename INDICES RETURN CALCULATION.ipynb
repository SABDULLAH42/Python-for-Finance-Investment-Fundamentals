```python
import numpy as np
import pandas as pd
from pandas_datareader import data as pdr
import matplotlib.pyplot as plt
import yfinance as yfin
```

```python
yfin.pdr_override()
```

```python
tickers = ['^GSPC', '^FTSE','^GDAXI','^IXIC']

ind_data = pd.DataFrame()

for t in tickers:

    ind_data[t] = yfin.download(t,start='1997-1-1')['Adj Close']
    
```

```python
ind_data.head()
```

```python
ind_data.tail()
```

```python
(ind_data/ind_data.iloc[0] * 100).plot(figsize = (15,6))
plt.show()
```

```python
ind_returns  = (ind_data/ ind_data.shift(1)) - 1
ind_returns.tail()
```

```python
annual_ind_returns = ind_returns.mean() * 250
annual_ind_returns
```
