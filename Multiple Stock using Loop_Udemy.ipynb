```python
import pandas as pd
from pandas_datareader import data as pdr
import yfinance as yfin


yfin.pdr_override()

PG = pdr.get_data_yahoo('PG', start='1995-1-1')
PG.head(10)
```

```python
tickers = ['PG', 'MSFT','F','GE']

mydata = pd.DataFrame()

for t in tickers:

    mydata[t] = yfin.download(t,start='1995-1-1')['Adj Close']
    
mydata.tail()
```
