```python
import yfinance as yf
import matplotlib.pyplot as plt
from datetime import datetime
%matplotlib inline
```

```python
start_date_str = input("Enter the start date (YYYY-MM-DD): ")
```

```python
end_date_str = input("Enter the end date (YYYY-MM-DD): ")
```

```python
start_date = datetime.strptime(start_date_str, '%Y-%m-%d')  ##Convert the input strings to datetime objects
end_date = datetime.strptime(end_date_str, '%Y-%m-%d')
```

```python
symbol = input("Enter the stock ticker symbol: ")
```

```python
data = yf.download(symbol, start=start_date, end=end_date) ##Get the data from Yahoo Finance API
```

```python
plt.plot(data['Close'])
plt.title(f"{symbol} Closing Prices ({start_date_str} to {end_date_str})")
plt.xlabel("Date")
plt.ylabel("Closing Price")
```

```python
plt.savefig(f"{symbol}_last_month_closing_prices.png", dpi=300)
```
